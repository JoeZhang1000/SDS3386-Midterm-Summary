## Week 1: Introduction

In this section, we go over the basics of MatPlotLib, Numpy, and Pandas.

### Matplotlib

Matplotlib is a library for creating data visualization. All we need to do is to pass the data to the plot function to create a plot.

```python
# Importing matplotlib

import matplotlib.pyplot as plt

# Generate data
temp = [28, 29, 13, 17, 14.5, 30, 21]

# Plot the data
plt.plot(temp)
```

Now, we have plotted our data. We now want to compute the mean and standard deviation of the data. We can do this the old-schooled way.

```python
mean = sum(temp)/len(temp)
print(mean)
```

This works. However, there's another package that can help do such operations with built-in functions. The package is called numpy, which we'll see below.

### Numpy

The numpy package contains built-in functions that can help complete mathematical operations such as computing statistics. For example, we use the data generated above and we compute the mean and standard deviation.

```python
# Importing numpy
import numpy as np

# Calculate mean and standard deviation

mean2 = np.mean(temp)
print(mean2)

s = np.std(temp)
print(s)
```

### Using Datasets

Now that we know some basic operations, we can now use a full dataset. We want to import data from csv files and conduct data analysis. We can use numpy to do this.

```python
# Import numpy package
import numpy as np

file = open('mpg_extract.csv', 'r') # Open file for reading
data = np.loadtxt(file, delimiter = ",")
print(data)
```

```python
# Now, we use matplotlib to see how engine size and mileage are related

import matplotlib.pyplot as plt

plt.plot(data[:,0], data[:,3]) # plot the first and fourth columns
```

```python
# We don't want there to be lines, so we do the following:

plt.scatter(data[:,0], data[:, 3])
```

```python
# We can also add labels to the chart

plt.scatter(data[:,0], data[:, 3])
plt.xlabel("displacement (engine size in litre)")
plt.ylabel("mileage (mpg)")
```

Now, we construct a plot for year vs. mileage

```python
# Plot year vs. mileage
plt.scatter(data[:,1], data[:, 3])
plt.xlabel("year")
plt.ylabel("mileage(mpg)")
```

This plot doesn't look ideal, so there's another way in which we can represent the year. We can do this by using colors and matplotlib allows us to do it.

```python
plt.scatter(data[:, 0], data[:, 3], c = data[:,1])
plt.xlabel("displacement (engine size in litre)")
plt.ylabel("mileage(mpg)")
```

```python
# We can also try a different colormap

plt.scatter(data[:,0], data[:,3], c = data[:, 1], cmap = 'coolwarm')
plt.xlabel('displacement(engine size in litre)')
plt.ylabel('mileage (mpg)')
```

We can also illustrate the number of cylinders in a similar way.

```python
plt.scatter(data[:, 0], data[:,3], c=data[:, 2], cmap = 'coolwarm')
plt.xlabel('displacement (engine size in litre)')
plt.ylabel('mileage (mpg)')
```

We have used color to seaprately represent two variables: year and number of cylinders. We want to illustrate both of them on the chart. To do this, we can use shapes. 

```python
group1 = data[np.where(data[:,1] == 1999)]
group2 = data[np.where(data[:,1] == 2008)]

plt.scatter(group1[:,0], group1[:,3], c = group1[:,2], cmap='coolwarm', marker = 'x')
plt.scatter(group2[:,0], group2[:,3], c = group2[:,2], cmap='coolwarm', marker = 'o')

plt.xlabel("displacement (engine size in litre)")
plt.ylabel("mileage (mpg)")
```

Matplotlib is an extremely versatile plotting package. We have used the "state-based" interface here. There is also an object-oriented interface. Many things can be done both ways, but some more complex things can be done via the object-oriented interface. 

See: https://matplotlib.org/matplotblog/posts/pyplot-vs-object-oriented-interface/
