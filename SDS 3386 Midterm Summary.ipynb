In this section, we go through how to construct data visualizations using the Altair package.

### Marks

There are multiple ways where we can construct plots. We use marks for that:

- `mark_circle` (points)
- `mark_point` (with variable shape)
- `mark_line` (for line charts)
- `mark_bar` (fot bar charts)

There are many more: https://altair-viz.github.io/altair-viz-v4/user_guide/marks.html

### Encodings

Simple:

```
x = 'Horsepower'
y = 'Miles_per_Gallon',
color = 'Origin'
```

Fancy:

```
x = alt.X('Horsepower:Q').axis(tickMinStep=50),
y = alt.Y('Miles_per_Gallon:Q).title('Miles per Gallon'),
color = 'Origin:N'
```

See: https://altair-viz.github.io/user_guide/encodings/index.html

````python
import warnings
warnings.filterwarnings('ignore')
````

````python
# We first import altair and some data
import altair as alt
from vega_datasets import data

# We first load some data
barley = data.barley() # From Altair documentation

# Construct a bar chart
alt.Chart(barley).mark_bar().encode(
    x=alt.X('variety:N'),
    y=alt.Y('sum(yield):Q'),
    color = 'site:N',
    order = alt.Order("site").sort("descending")
)
````

Stacked bar charts are good at showing a total and proportion for each category. However, they are not very good at showing the value for each category or allowing easy comparison of values between categories.

### Altair has many other features, including

- setting axis scales
- aggregating and filtering data
- combining plots
- interactivity

````python
# Create a line chart

source = data.population() # From Altair documentation

alt.Chart(source).mark_line().encode(
    x = 'year:O',
    y = alt.Y('sum(people)', scale=alt.Scale(type = 'log')) # log scale
)
````

````python
# Modified version that changes the y axis label and uses method-based syntex

source = data.population()

alt.Chart(source).mark_line().encode(
    x = 'year:O',
    y = alt.Y('sum(people)').title('Population').scale(type = 'log')
)
````

````python
# Aggregation by mean

cars = data.cars.url

chart = alt.Chart(cars).mark_bar().encode(
    y = 'Cylinders:O',
    x = 'mean_acc:Q'
).transform_aggregate(
    mean_acc = 'mean(Acceleration)',
    groupby=['Cylinders']
)

chart
````

Mysteriously, Altair has decided to shade different dots in different colours. This is because Altair's goal is to automatically choose useful plot settings and configurations so the user is free to think about the data rather than the mechanics of plotting. In the case above, it is because the dark dots represent the auto cars and the light dots represent the manual cars.

We can fine tune the chart. The chart produced using Pandas has larger circles. The reason is because the chart produced by Altair always starts at 0 (this is always the case with Altair charts), which can help avoid perceptual errors.

To get the best of both worlds, we can use Altair's interactive mode. We'll also change the horizontal scale to go up to 7.5 so we don't get a circle on the right edge of the plot. We'll also add a tooltip and a title.

We also experimented with making the color blue and use mark_point instead of mark_circle.

````python
alt.Chart(mpg, title = "Larger engines generally have poorer fuel economy").encode(
    x = alt.X('displ', scale=alt.Scale(domain=[1,7.5])),
    y = 'hwy',
    tooltip = 'model',
    color = alt.value('blue')
).mark_point(size=40).interactive()
````

Numpy array slicing produces sub-array that is just a view of the array, not a copy (by default). Modifying the sub-array also modifies the original array.

````python
a = np.array([[1,2,3.], [4,5,6]])

sub1 = a[1, 0:-1]
print(sub1)

sub1[0] = 13
print(sub1)

print(a)
````

We can force a copy with the function `copy()` as follows:

````python
# Forcing a copy:
a = np.array([[1,2,3.], [4,5,6]])

sub1 = a[1,0:-1].copy()
print(sub1)

sub1[0] = 13
print(sub1)

print(a)
````

As you can see, the original array is not modified if you force a copy and modify it.

### Data Cleaning

We want our data to be tidy. A dataframe is tidy if both of the following are true:
- Each column corresponds to a variable
- Each row corresponds to an observation

These must be one-to-one correspondences.

It follows that each cell contains the observed value of a variable. This structure may seem bovious, but it is surprisingly rare in the wild.

It may be hard to decide what is a variable. However, there is a rule of thumb: If it makes sense to compare two values, then they are observations of the same variable.

For example, given two columns "Asst1" and "Asst2" containing grades, are these one variable or two variables?

If Asst1 and Asst2 are both on the same scale, then they can be compared, so they are best considered as the same variable. Assignment number becomes another variable.

We can tidy dataframes using pivot functions. These functions are `pd.pivot`, `pd.wide_to_long`, or `pd.melt`.

If the assignment grades are not on the same scale, then it is better to rescale them so they become comparable.

In a tidy dataframe, columns are of two types:
- Keys/idenfifying variables/ID variables/independent variables
- Measures/dependent values

No two rows have the same combination of all identifying variables.

A tuple of key values that determines a unique row is called a multi-index

In this section, we talk about probability and statistics and how we can do calculations using python commands.

### Review of Probability

Let $A$ and $B$ be events. We say that $A$ and $B$ are disjoint if they can't both happen at the same time, or $P(A \cap B) = 0$. 

If $A$ and $B$ are disjoint then $P(A \cup B) = P(A) + P(B)$.

$A$ and $B$ are independent if and only if $P(A \cap B) = P(A)P(B)$

The conditional probability of $A$ given $B$ is denoted $P(A|B)$ and defined by $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

Example: A fair 6-sided die is rolled. What is the probability that the outcome is 4, given that the outcome is even?

Solution:

Let $X$ be the outcome. Then $P(X = 1) = \dots = P(X = 6) = \frac{1}{6}$. We can then calculate the probability as follows: 

\begin{align}
P(X \ is \ even) 
&= P(X = 2 \ or \ X = 4 \ or X = 6) \\
&= P(X = 2) + P(X = 4) + P(X = 6) = \frac{1}{2}.
\end{align}

<br>

So
$$
P(X = 4 \mid X \ is \ even) = \frac{P(X = 4)}{P(X \ is \ even)} 
= \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}.
$$

### Bayes' Theorem

Now we derive Bayes' Theorem using the definition of conditional probabilty.

The definition of conditional probability is $$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

This can be rearranged as $$P(A \cap B) = P(A | B)P(B)$$

We also have that $$P(A \cap B) = P(B \cap A) = P(B | A)P(A)$$

Then we get $$P(A|B)P(B) = P(B|A)P(A)$$

So now, if we rearrange the equation above, we get Bayes' Theorem:

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$

When applying this formula and computing $P(B)$, it is often useful to decompose it as 

$$P(B) = P(B \cap A) + P(B \cap \overline{A}) = P(B|A)P(A) + P(B|A)P(\overline{A})$$

Similarly, if there are disjoint events $A_1, ..., A_n$ with $P(A_1) + \dots + P(A_n) = 1$, hten

$$P(B) = P(B \cap A_1) + \dots + P(B \cap A_n)$$

Example: Consider a random screening program for a disease that affects 1 in 500 people in the general population. Suppose the screening test has a sensitivity of 99\%, meaning that the test is positive 99\% of the time in people who actually have the disease; and a specificity of 98\%, meaning that when people who don't have the disease are tested, the test gives a false positive result 2\% of the time. If someone tests positive, what is the probability that they actually have the disease?

We define the following events:

$A$: Event that person has the disease
$B$: Event that the test is positive

We also have the following probabilities:

$P(A) = \frac{1}{500}$
$P(B|A) = 0.99$
$P(B|\overline{A}) = 0.02$

We want to now find $P(A|B)$. To do that, we need to find $P(B)$ first

$$P(B) = P(B|A)P(A) + P(B|\overline{A})P(\overline{A}) = (0.99)(\frac{1}{500}) + (0.02)(\frac{499}{500}) = 0.02194$$

Now, we find the probability

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)} = \frac{(0.99)(\frac{1}{500})}{0.02194} = 0.09$$

Therefore, if a person tests positive, there's a 0.09 probability that they actually have the disease.

There are some known probability distributions. Discrete probability distributions on a finite set are determined by probability mass functions that define the probability for each value. 

E.g. The Uniform distribution on $\{1, 2, ..., n\}$ has p.m.f. $$p(x) = \frac{1}{n}$$ for all $x \in \{1, 2, ..., n\}$

E.g. The Bernoulli(p) distribution on $\{0, 1\}$ has p.m.f. $$p(1) = p$$ $$p(0) = 1-p$$

E.g. The Binomial(n, p) distribution on $\{0,1,2,...,n\}$ has p.m.f. $$p(k) = {n \choose k}p^k(1-p)^{n-k}$$

Now, we talk about some statistics for looking at pairs of observations.

Given pairs of observations of variables $X$ and $Y$, $$(x_1, y_1), ..., (x_n, y_n)$$

the ***sample covariance*** is: $$cov(X, Y) := \frac{1}{n} \sum_{i=1}^{n} (x_i - m_x)(y_i - m_y)$$ where $m_x$ and $m_y$ are the sample means of the two variables.

Note that 
- $cov(X,Y) = cov(Y,X)$
- $cov(X,X) = var(X)$

In general, covariance measures how much two variables vary with each other.

If $cov(X, Y) > 0$ then $X$ and $Y$ are positively associated.

If $cov(X, Y) < 0$ then $X$ and $Y$ are negatively associated.

Given a sequence of variables $(X_1, ..., X_p)$, the ***covariance matrix*** is the $p \times p$ matrix $Cov(X_1, X_2, ..., X_p)$ such that the $(i, j)$th entry is $Cov(X_i, X_j)$. The diagonal entries are $Var(X_i)$.

````python
# Example for covariance matrix

mass = pen['body_mass_g']
fliplen = pen['flipper_length_mm']

# variances, computed individually:
varm = pen['body_mass_g'].var()
varf = pen['flipper_length_mm'].var()
print(f'Variance of body mass: {varm:.1f}; variance of flipper length: {varf:.1f}')
# Neither numpy nor pandas can compute individual covariances, but they both have a function that computes the covariance matrix
pen[['body_mass_g', 'flipper_length_mm']].cov()
````

There is a problem of using covariance to measure how 2 variables vary with each other. It is scale-dependent, which makes it hard to compare covariance values between different pairs of variables.

This motivates the ***correlation*** between two variables: $$corr(X, Y) = \frac{cov(X, Y)}{std(X)std(Y)}$$ where std is standard deviation.

Like covariance, a positive correlation denotes a postivie association. However, the correlation is always between -1 and 1. 

Correlation is comparable between pairs of variables, and its absolute value indicates the strength of correlation.

````python
# Example of correlation
pen[['body_mass_g', 'flipper_length_mm']].corr()
````
